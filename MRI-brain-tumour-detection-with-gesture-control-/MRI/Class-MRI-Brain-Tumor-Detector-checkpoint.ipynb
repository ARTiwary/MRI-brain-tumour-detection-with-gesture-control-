# Visualizing brain MRI images

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_classes(glioma, meningioma, pituitary, notumor, num=5):
    classes = {
        "Glioma": glioma,
        "Meningioma": meningioma,
        "Pituitary": pituitary,
        "No Tumor": notumor
    }
    
    plt.figure(figsize=(15, 10))
    
    for row, (class_name, imgs) in enumerate(classes.items()):
        indices = np.random.choice(len(imgs), num, replace=False)
        
        for col, idx in enumerate(indices):
            plt.subplot(len(classes), num, row * num + col + 1)
            plt.imshow(imgs[idx])
            plt.title(class_name)
            plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Call it
visualize_classes(glioma, meningioma, pituitary, notumor, num=5)
```

# Create Torch Dataset Class


## Abstract Dataset Class


```python
from torch.utils.data import ConcatDataset

class Dataset(object):
    """
    An abstract base class representing a Dataset.

    All datasets should subclass this and override:
        - __len__: returns the size of the dataset
        - __getitem__: supports integer indexing from 0 to len(self)-1
    """

    def __getitem__(self, index):
        """
        Retrieve the data sample at the given index.
        Must be implemented by subclasses.
        """
        raise NotImplementedError("Subclasses must implement __getitem__")

    def __len__(self):
        """
        Return the total number of samples in the dataset.
        Must be implemented by subclasses.
        """
        raise NotImplementedError("Subclasses must implement __len__")

    def __add__(self, other):
        """
        Concatenate this dataset with another dataset.
        Returns a ConcatDataset object from PyTorch.
        """
        if not isinstance(other, Dataset):
            raise TypeError("Can only concatenate with another Dataset")
        return ConcatDataset([self, other])
```

# Creating MRI custom dataset class

```python
import glob
import cv2
import numpy as np
import torch
from torch.utils.data import Dataset

class MRI(Dataset):
    """
    Custom PyTorch Dataset for 4-class Brain Tumor MRI images:
    0 = Glioma, 1 = Meningioma, 2 = Pituitary, 3 = No Tumor
    """

    def __init__(self, data_dirs, labels, image_size=(128, 128), transform=None):
        """
        data_dirs: list of folder paths for each class, e.g.,
                   ['./data/Training/glioma', './data/Training/meningioma', ...]
        labels: list of integer labels corresponding to each folder, e.g., [0, 1, 2, 3]
        image_size: resize images to (height, width)
        transform: optional image transformations (augmentations)
        """
        self.images = []
        self.labels = []
        self.transform = transform
        self.image_size = image_size

        for folder, label in zip(data_dirs, labels):
            for f in glob.iglob(folder + "/*.jpg"):
                img = cv2.imread(f)
                img = cv2.resize(img, image_size)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # BGR → RGB
                img = img.astype(np.float32)
                # Convert HWC → CHW for PyTorch
                img = np.transpose(img, (2, 0, 1))
                self.images.append(img)
                self.labels.append(label)

        self.images = np.array(self.images, dtype=np.float32)
        self.labels = np.array(self.labels, dtype=np.int64)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, index):
        image = self.images[index]
        label = self.labels[index]

        if self.transform:
            image = self.transform(image)

        # Convert to PyTorch tensors
        image = torch.tensor(image, dtype=torch.float32)
        label = torch.tensor(label, dtype=torch.long)

        return {'image': image, 'label': label}

    def normalize(self):
        """Normalize images to [0,1] range."""
        self.images = self.images / 255.0
data_dirs = [
    "./data/Training/glioma",
    "./data/Training/meningioma",
    "./data/Training/pituitary",
    "./data/Training/notumor"
]

labels = [0, 1, 2, 3]

# Create dataset
dataset = MRI(data_dirs, labels)

# Normalize images
dataset.normalize()

# Test first sample
sample = dataset[0]
print("Image shape:", sample['image'].shape) 
print("Label:", sample['label'])
```

# Creating a dataloader

```python
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

names = {0: "Glioma", 1: "Meningioma", 2: "Pituitary", 3: "No Tumor"}
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)

for i, sample in enumerate(dataloader):
    img = sample['image'].squeeze().permute(1,2,0).numpy()  # CHW -> HWC
    plt.imshow((img*255).astype(np.uint8))
    plt.title(names[sample['label'].item()])
    plt.axis('off')
    plt.show()
    
    if i == 5:  # show first 6 images
        break
```
