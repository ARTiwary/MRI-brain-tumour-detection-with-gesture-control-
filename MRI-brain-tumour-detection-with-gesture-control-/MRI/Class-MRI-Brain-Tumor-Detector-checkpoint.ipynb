# Create Torch Dataset Class


## Abstract Dataset Class


```python
from torch.utils.data import ConcatDataset

class Dataset(object):
    """
    An abstract base class representing a Dataset.

    All datasets should subclass this and override:
        - __len__: returns the size of the dataset
        - __getitem__: supports integer indexing from 0 to len(self)-1
    """

    def __getitem__(self, index):
        """
        Retrieve the data sample at the given index.
        Must be implemented by subclasses.
        """
        raise NotImplementedError("Subclasses must implement __getitem__")

    def __len__(self):
        """
        Return the total number of samples in the dataset.
        Must be implemented by subclasses.
        """
        raise NotImplementedError("Subclasses must implement __len__")

    def __add__(self, other):
        """
        Concatenate this dataset with another dataset.
        Returns a ConcatDataset object from PyTorch.
        """
        if not isinstance(other, Dataset):
            raise TypeError("Can only concatenate with another Dataset")
        return ConcatDataset([self, other])
```

# Creating MRI custom dataset class

```python
import glob
import cv2
import numpy as np
import torch
from torch.utils.data import Dataset

class MRI(Dataset):
    """
    4-class Brain Tumor MRI Dataset
    0 = Glioma, 1 = Meningioma, 2 = Pituitary, 3 = No Tumor
    """

    def __init__(self, data_dirs, labels, image_size=(128, 128), transform=None):
        self.images = []
        self.labels = []
        self.transform = transform

        for folder, label in zip(data_dirs, labels):
            for f in glob.iglob(folder + "/*.jpg"):
                img = cv2.imread(f)

                if img is None:
                    continue

                img = cv2.resize(img, image_size)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                # Normalize
                img = img.astype(np.float32) / 255.0

                # HWC → CHW
                img = np.transpose(img, (2, 0, 1))

                self.images.append(img)
                self.labels.append(label)

        self.images = np.array(self.images, dtype=np.float32)
        self.labels = np.array(self.labels, dtype=np.int64)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, index):
        image = self.images[index]
        label = self.labels[index]

        if self.transform:
            image = np.transpose(image, (1, 2, 0))  # CHW → HWC
            image = self.transform(image)
        else:
            image = torch.from_numpy(image).float()

        label = torch.tensor(label, dtype=torch.long)

        return {'image': image, 'label': label}

# TRAIN + TEST DATASET

train_dirs = [
    "./data/Training/glioma",
    "./data/Training/meningioma",
    "./data/Training/pituitary",
    "./data/Training/notumor"
]

test_dirs = [
    "./data/Testing/glioma",
    "./data/Testing/meningioma",
    "./data/Testing/pituitary",
    "./data/Testing/notumor"
]

labels = [0, 1, 2, 3]

train_dataset = MRI(train_dirs, labels)
test_dataset = MRI(test_dirs, labels)

# TEST SAMPLE

sample = train_dataset[0]
print("Image shape:", sample['image'].shape)
print("Label:", sample['label'])
```

# Create a model

```python
import torch.nn as nn

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        
        self.cnn_model = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(16),
            nn.MaxPool2d(2),

            
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(32),
            nn.MaxPool2d(2),

            
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            nn.MaxPool2d(2)
        )

        self.fc_model = nn.Sequential(
            nn.Linear(64 * 16 * 16, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 4)  
        )

    def forward(self, x):
        x = self.cnn_model(x)
        x = x.view(x.size(0), -1)
        x = self.fc_model(x)
        return x
```

# numpy array vs tensor
The difference between a Numpy array and a tensor is that the tensor are backed by the accelerator memory like GPU and they are immutable, unlike NumPy arrays. You can never update a tensor but create a new one. If you are into machine learning or going to be into it , A tensor is a suitable choice if you are going to use **GPU**. A tensor can reside in accelerator's memory.

* The numpy arrays are the core functionality of the numpy package designed to support faster mathematical operations. Unlike python's inbuilt list data structrure, they can only hold elements of a  single data type. Library like pandas which is used for data preprocessing is built around the numpy array. **Pytorch tensors are similar to numpy array, but can also be operated on CUDA-capable NVIDIA GPU**
* Numpy arrays are mainly used in typicaly machine learning algorithms (such as k-means or Decision Tree in scikit-learn) whereas pytorch tensors are mainly used in deep learing which require heavy matrix computation.
* Unlike numpy arrays, while creating pytorch tensor, it also accepts two other arguments called the device _type (weather the computation happens on CPU or GPU) and the require_grad (which is used to computer the derivatives).

# Train the Dumb Model

```python
# Hyperparameters
EPOCH = 50
LR = 0.001
BATCH_SIZE = 32

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Optimizer and loss
optimizer = torch.optim.Adam(model.parameters(), lr=LR)
criterion = nn.CrossEntropyLoss()

# Training loop
for epoch in range(1, EPOCH + 1):
    model.train()
    epoch_losses = []
    correct = 0
    total = 0
    
    for batch in train_loader:
        images = batch['image'].to(device)
        labels = batch['label'].to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        epoch_losses.append(loss.item())
        preds = torch.argmax(outputs, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
    
    # Print train stats
    if epoch % 5 == 0 or epoch == 1:
        print(f"Epoch [{epoch}/{EPOCH}] | Loss: {np.mean(epoch_losses):.4f} | Accuracy: {correct/total:.4f}")


# Evaluate on test set after training
model.eval()
y_true = []
y_pred = []

with torch.no_grad():
    for batch in test_loader:
        images = batch['image'].to(device)
        labels = batch['label'].to(device)
        outputs = model(images)
        preds = torch.argmax(outputs, dim=1)
        
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())

print("\nTest Accuracy:", np.mean(np.array(y_true) == np.array(y_pred)))
```

**Are We Overfitting?**

To check whether our model is overfitting, we introduce a validation set.
This allows us to compare performance on:

Training data (what the model learns from)
Validation data (unseen data during training)

```python
import glob, cv2, numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from torchvision import transforms

# --- Dataset ---
class MRI(Dataset):
    def __init__(self, mode='train', transform=None):
        self.mode = mode
        self.transform = transform

        # Load images
        self.images, self.labels = [], []
        classes = ['glioma','meningioma','pituitary','notumor']
        for idx, cls in enumerate(classes):
            for f in glob.iglob(f"./data/Training/{cls}/*.jpg"):
                img = cv2.imread(f)
                if img is None: continue
                img = cv2.cvtColor(cv2.resize(img,(128,128)), cv2.COLOR_BGR2RGB)
                self.images.append(img)
                self.labels.append(idx)
        
        self.images = np.array(self.images, dtype=np.float32) / 255.0  # normalize to [0,1]
        self.labels = np.array(self.labels, dtype=np.int64)

        # Split train/val
        X_train, X_val, y_train, y_val = train_test_split(
            self.images, self.labels, test_size=0.2, random_state=42, stratify=self.labels
        )
        self.X_train, self.X_val = X_train, X_val
        self.y_train, self.y_val = y_train, y_val

    def __len__(self):
        return len(self.X_train) if self.mode=='train' else len(self.X_val)

    def __getitem__(self, idx):
        image = self.X_train[idx] if self.mode=='train' else self.X_val[idx]
        label = self.y_train[idx] if self.mode=='train' else self.y_val[idx]
        if self.transform:
            image = self.transform(image)
        else:
            image = torch.tensor(image).permute(2,0,1)  # HWC -> CHW
        label = torch.tensor(label)
        return image, label

# --- Transforms ---
train_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15)
])

val_transform = transforms.Compose([
    transforms.ToTensor()
])

# --- DataLoaders ---
train_dataset = MRI(mode='train', transform=train_transform)
val_dataset = MRI(mode='val', transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

print(f"Number of training samples: {len(train_dataset)}")
print(f"Number of validation samples: {len(val_dataset)}")
```

```python
import torch.nn as nn
import torch.optim as optim

# Example simple CNN (replace with your own)
class SimpleCNN(nn.Module):
    def __init__(self, num_classes=4):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3,32,3,padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32,64,3,padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(64,128,3,padding=1), nn.ReLU(), nn.MaxPool2d(2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128*16*16, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256,num_classes)
        )
    def forward(self,x):
        x = self.features(x)
        x = self.classifier(x)
        return x

device = "cuda" if torch.cuda.is_available() else "cpu"
model = SimpleCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# --- Training ---
best_val_loss = float('inf')
patience = 10
trigger = 0

for epoch in range(1,101):
    model.train()
    train_loss, correct, total = 0,0,0
    for imgs, labels in train_loader:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()*imgs.size(0)
        total += labels.size(0)
        correct += (outputs.argmax(1)==labels).sum().item()
    train_loss /= total
    train_acc = correct/total

    # Validation
    model.eval()
    val_loss, val_correct, val_total = 0,0,0
    with torch.no_grad():
        for imgs, labels in val_loader:
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            val_loss += loss.item()*imgs.size(0)
            val_total += labels.size(0)
            val_correct += (outputs.argmax(1)==labels).sum().item()
    val_loss /= val_total
    val_acc = val_correct/val_total

    print(f"Epoch {epoch}/100 | Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")

    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        trigger = 0
        torch.save(model.state_dict(),"best_model.pth")
    else:
        trigger += 1
        if trigger >= patience:
            print(f"Early stopping at epoch {epoch}")
            break
```
