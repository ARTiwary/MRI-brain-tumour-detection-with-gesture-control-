# Create Torch Dataset Class


## Abstract Dataset Class


```python
from torch.utils.data import ConcatDataset

class Dataset(object):
    """
    An abstract base class representing a Dataset.

    All datasets should subclass this and override:
        - __len__: returns the size of the dataset
        - __getitem__: supports integer indexing from 0 to len(self)-1
    """

    def __getitem__(self, index):
        """
        Retrieve the data sample at the given index.
        Must be implemented by subclasses.
        """
        raise NotImplementedError("Subclasses must implement __getitem__")

    def __len__(self):
        """
        Return the total number of samples in the dataset.
        Must be implemented by subclasses.
        """
        raise NotImplementedError("Subclasses must implement __len__")

    def __add__(self, other):
        """
        Concatenate this dataset with another dataset.
        Returns a ConcatDataset object from PyTorch.
        """
        if not isinstance(other, Dataset):
            raise TypeError("Can only concatenate with another Dataset")
        return ConcatDataset([self, other])
```

# Creating MRI custom dataset class

```python
class MRI(Dataset):
    def __init__(self, data_dirs, labels, image_size=(128, 128), transform=None):
        self.images = []
        self.labels = []
        self.transform = transform
        self.image_size = image_size

        for folder, label in zip(data_dirs, labels):
            for f in glob.iglob(folder + "/*.jpg"):
                img = cv2.imread(f)
                if img is None:
                    continue
                img = cv2.resize(img, image_size)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                # Store as uint8 HWC — DO NOT normalize here
                self.images.append(img)
                self.labels.append(label)

        self.labels = np.array(self.labels, dtype=np.int64)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, index):
        image = self.images[index]   # uint8 HWC numpy
        label = self.labels[index]

        if self.transform:
            image = self.transform(image)   # → CHW float tensor
        else:
            image = torch.from_numpy(
                image.transpose(2, 0, 1).astype(np.float32) / 255.0
            )

        return {'image': image, 'label': torch.tensor(label, dtype=torch.long)}

# ─────────────────────────────────────────
# LOAD DATASETS WITH TRANSFORMS
# ─────────────────────────────────────────
train_dirs = [
    "./data/Training/glioma",
    "./data/Training/meningioma",
    "./data/Training/pituitary",
    "./data/Training/notumor"
]
test_dirs = [
    "./data/Testing/glioma",
    "./data/Testing/meningioma",
    "./data/Testing/pituitary",
    "./data/Testing/notumor"
]
labels = [0, 1, 2, 3]

# Pass transforms here!
train_dataset = MRI(train_dirs, labels, transform=train_transform)
test_dataset  = MRI(test_dirs,  labels, transform=test_transform)

print(f"Train: {len(train_dataset)} | Test: {len(test_dataset)}")
```

```python
def unnormalize(img_tensor):
    """Convert normalized tensor back to displayable [0,1] image."""
    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])
    img = img_tensor.permute(1, 2, 0).cpu().numpy()  # CHW → HWC
    img = img * std + mean                            # undo normalization
    img = np.clip(img, 0, 1)                          # clip to valid range
    return img
```

# Create a model

```python

class CNN(nn.Module):
    def __init__(self, num_classes=4):
        super(CNN, self).__init__()

        self.features = nn.Sequential(
            # Block 1 — 128×128 → 64×64
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.1),

            # Block 2 — 64×64 → 32×32
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.1),

            # Block 3 — 32×32 → 16×16
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.2),

            # Block 4 — 16×16 → 8×8
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.2),
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),                          # 256 * 8 * 8 = 16384
            nn.Linear(256 * 8 * 8, 512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(512, 128),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x


# ─────────────────────────────────────────
# CREATE MODEL
# ─────────────────────────────────────────
model = CNN(num_classes=4)

# ─────────────────────────────────────────
# SANITY CHECK — dummy forward pass
# ─────────────────────────────────────────
dummy = torch.randn(1, 3, 128, 128)
with torch.no_grad():
    out = model(dummy)

print("=" * 40)
print("        MODEL CREATED ✓")
print("=" * 40)
print(f"Input  shape : {dummy.shape}")
print(f"Output shape : {out.shape}")        # should be [1, 4]
print(f"Num classes  : {out.shape[1]}")
print("=" * 40)
print(model)
print("=" * 40)

# Total parameters
total_params     = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total params     : {total_params:,}")
print(f"Trainable params : {trainable_params:,}")
print("=" * 40)
```

# numpy array vs tensor
The difference between a Numpy array and a tensor is that the tensor are backed by the accelerator memory like GPU and they are immutable, unlike NumPy arrays. You can never update a tensor but create a new one. If you are into machine learning or going to be into it , A tensor is a suitable choice if you are going to use **GPU**. A tensor can reside in accelerator's memory.

* The numpy arrays are the core functionality of the numpy package designed to support faster mathematical operations. Unlike python's inbuilt list data structrure, they can only hold elements of a  single data type. Library like pandas which is used for data preprocessing is built around the numpy array. **Pytorch tensors are similar to numpy array, but can also be operated on CUDA-capable NVIDIA GPU**
* Numpy arrays are mainly used in typicaly machine learning algorithms (such as k-means or Decision Tree in scikit-learn) whereas pytorch tensors are mainly used in deep learing which require heavy matrix computation.
* Unlike numpy arrays, while creating pytorch tensor, it also accepts two other arguments called the device _type (weather the computation happens on CPU or GPU) and the require_grad (which is used to computer the derivatives).

# torch.tensor vs torch.cuda.tensor
The key difference is just that **torch.tensor** occpuies **CPU** memory  while **torch.cuda.tensor** occpuies GPU memory.

As the underlying hardware interface is completly different, CPU Tensor are just compatible with CPU Tensor and GPU Tensor are just compatible with GPU Tensors.

In which scenario is **torch.cuda.Tensor()** necessary ?
When you want to use GPU accleration (which is much faster in most cases) for your program, you need to use **torch.cuda.Tensor**, but you have to make sure that all tensors you are using are **CUDA Tensors**, mixing is not possible here.

**tensor.cpu().detach().numpy()** : Convert PyTorch tensor to NumPy array.
As mentioned before, **np.ndarray** objects do not have this extra "Computational graph" layer and therefore, when converting a torch.Tensor to np.ndarray, you must explicitly remove the computational graph of the tensor using the detach() command .cpu() return a copy of this object in CPU memory.

# Train the Dumb Model

```python
import torch
import torch.nn as nn
import numpy as np
from torch.utils.data import DataLoader, random_split
import matplotlib.pyplot as plt

EPOCH      = 50
LR         = 0.001
BATCH_SIZE = 32

train_size = int(0.8 * len(train_dataset))
val_size   = len(train_dataset) - train_size
train_data, val_data = random_split(train_dataset, [train_size, val_size])

train_loader = DataLoader(train_data,    batch_size=BATCH_SIZE, shuffle=True,  num_workers=0)
val_loader   = DataLoader(val_data,      batch_size=BATCH_SIZE, shuffle=False, num_workers=0)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, shuffle=False, num_workers=0)

device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model     = model.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=15, gamma=0.5)
criterion = nn.CrossEntropyLoss()

print("Device:", device)

# ─────────────────────────────────────────
# TRAIN LOOP
# ─────────────────────────────────────────
train_losses, val_losses = [], []
train_accs,   val_accs   = [], []

for epoch in range(1, EPOCH + 1):
    # ── Train ──
    model.train()
    epoch_loss, correct, total = [], 0, 0

    for batch in train_loader:
        images = batch['image'].to(device)
        labels = batch['label'].to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss    = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        epoch_loss.append(loss.item())
        correct += (outputs.argmax(1) == labels).sum().item()
        total   += labels.size(0)

    scheduler.step()

    # ── Validate ──
    model.eval()
    v_loss, v_correct, v_total = 0, 0, 0

    with torch.no_grad():
        for batch in val_loader:
            images = batch['image'].to(device)
            labels = batch['label'].to(device)
            outputs = model(images)
            v_loss    += criterion(outputs, labels).item()
            v_correct += (outputs.argmax(1) == labels).sum().item()
            v_total   += labels.size(0)

    t_loss = np.mean(epoch_loss)
    t_acc  = correct / total
    v_loss = v_loss / len(val_loader)
    v_acc  = v_correct / v_total

    train_losses.append(t_loss); val_losses.append(v_loss)
    train_accs.append(t_acc);   val_accs.append(v_acc)

    print(f"Epoch {epoch:02d}/{EPOCH} | "
          f"Train Loss: {t_loss:.4f} Acc: {t_acc:.4f} | "
          f"Val Loss: {v_loss:.4f} Acc: {v_acc:.4f}")
```

```python
import torch
import torch.nn as nn
import numpy as np
from torchvision import models
from torch.utils.data import DataLoader, random_split

# ─────────────────────────────────────────
# RESNET18 — pretrained backbone
# ─────────────────────────────────────────
def build_resnet(num_classes=4, freeze_backbone=False):
    model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

    # Optionally freeze early layers
    if freeze_backbone:
        for name, param in model.named_parameters():
            if not name.startswith('layer4') and not name.startswith('fc'):
                param.requires_grad = False

    # Replace final classifier
    in_features = model.fc.in_features        # 512
    model.fc = nn.Sequential(
        nn.Linear(in_features, 256),
        nn.ReLU(),
        nn.Dropout(0.4),
        nn.Linear(256, num_classes)
    )
    return model

model = build_resnet(num_classes=4, freeze_backbone=False).to(device)

# Verify output shape
dummy = torch.randn(1, 3, 128, 128).to(device)
with torch.no_grad():
    out = model(dummy)
print(f"Output shape : {out.shape}")   # [1, 4]
print("ResNet18 ready ✓")

# ─────────────────────────────────────────
# CLASS WEIGHTS — stronger Glioma penalty
# ─────────────────────────────────────────
class_weights = torch.tensor([
    5.0,   # Glioma      — still at 78%, needs hard push
    2.0,   # Meningioma  — at 94.25%, small boost
    1.0,   # Pituitary   — fine
    1.0,   # No Tumor    — fine
], dtype=torch.float).to(device)

criterion = nn.CrossEntropyLoss(weight=class_weights)

# ─────────────────────────────────────────
# DATALOADERS
# ─────────────────────────────────────────
train_size = int(0.8 * len(train_dataset))
val_size   = len(train_dataset) - train_size
generator  = torch.Generator().manual_seed(42)
train_data, val_data = random_split(train_dataset, [train_size, val_size],
                                     generator=generator)

train_loader = DataLoader(train_data,   batch_size=32, shuffle=True,  num_workers=0)
val_loader   = DataLoader(val_data,     batch_size=32, shuffle=False, num_workers=0)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=0)

# ─────────────────────────────────────────
# OPTIMIZER — two learning rates
# lower lr for pretrained backbone
# higher lr for new classifier head
# ─────────────────────────────────────────
backbone_params   = [p for n, p in model.named_parameters()
                     if not n.startswith('fc')]
classifier_params = [p for n, p in model.named_parameters()
                     if n.startswith('fc')]

optimizer = torch.optim.Adam([
    {'params': backbone_params,   'lr': 1e-4},   # fine-tune slowly
    {'params': classifier_params, 'lr': 1e-3},   # train head faster
], weight_decay=1e-4)

scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)

# ─────────────────────────────────────────
# TRAINING
# ─────────────────────────────────────────
EPOCH         = 40
best_val_loss = float('inf')
patience      = 8
trigger       = 0

train_losses, val_losses = [], []
train_accs,   val_accs   = [], []

resnet_save_path = save_path.replace(
    'tumor-detection-model.pth',
    'tumor-detection-resnet18.pth'
)

print("=" * 55)
print("   RESNET18 TRAINING")
print("   Glioma weight: 5.0 | Meningioma weight: 2.0")
print("=" * 55)

for epoch in range(1, EPOCH + 1):
    # ── Train ──
    model.train()
    epoch_loss, correct, total = [], 0, 0

    for batch in train_loader:
        images = batch['image'].to(device)
        labels = batch['label'].to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss    = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        epoch_loss.append(loss.item())
        correct += (outputs.argmax(1) == labels).sum().item()
        total   += labels.size(0)

    scheduler.step()

    # ── Validate ──
    model.eval()
    v_loss, v_correct, v_total = 0, 0, 0

    with torch.no_grad():
        for batch in val_loader:
            images = batch['image'].to(device)
            labels = batch['label'].to(device)
            outputs = model(images)
            v_loss    += criterion(outputs, labels).item()
            v_correct += (outputs.argmax(1) == labels).sum().item()
            v_total   += labels.size(0)

    t_loss = np.mean(epoch_loss)
    t_acc  = correct / total
    v_loss = v_loss / len(val_loader)
    v_acc  = v_correct / v_total

    train_losses.append(t_loss)
    val_losses.append(v_loss)
    train_accs.append(t_acc)
    val_accs.append(v_acc)

    print(f"Epoch {epoch:02d}/{EPOCH} | "
          f"Train Loss: {t_loss:.4f} Acc: {t_acc:.4f} | "
          f"Val Loss: {v_loss:.4f} Acc: {v_acc:.4f}")

    # ── Save best + early stopping ──
    if v_loss < best_val_loss:
        best_val_loss = v_loss
        trigger       = 0
        torch.save({
            'model_state_dict': model.state_dict(),
            'model_name':  'resnet18',
            'num_classes': 4,
            'class_names': {0:'Glioma',1:'Meningioma',2:'Pituitary',3:'No Tumor'},
            'input_size':  (3, 128, 128),
            'val_acc':     round(v_acc * 100, 2),
            'train_acc':   round(t_acc * 100, 2),
            'epoch':       epoch,
            'val_loss':    round(v_loss, 4),
        }, resnet_save_path)
        print(f"  ✓ Saved (val loss → {v_loss:.4f})")
    else:
        trigger += 1
        if trigger >= patience:
            print(f"\nEarly stopping at epoch {epoch}")
            break

# ─────────────────────────────────────────
# LOAD BEST + EVALUATE
# ─────────────────────────────────────────
checkpoint = torch.load(resnet_save_path, map_location=device)
model.load_state_dict(checkpoint['model_state_dict'])
model.eval()
print(f"\nLoaded best — epoch {checkpoint['epoch']} "
      f"val acc {checkpoint['val_acc']}%")

y_true, y_pred, y_probs = [], [], []
with torch.no_grad():
    for batch in test_loader:
        images  = batch['image'].to(device)
        labels  = batch['label'].to(device)
        outputs = model(images)
        probs   = torch.softmax(outputs, dim=1)
        preds   = torch.argmax(outputs, dim=1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())
        y_probs.extend(probs.cpu().numpy())

y_true  = np.array(y_true)
y_pred  = np.array(y_pred)
y_probs = np.array(y_probs)

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print("\n" + "=" * 55)
print("   FINAL SENSITIVITY & SPECIFICITY — RESNET18")
print("=" * 55)
print(f"\n  Overall Test Accuracy: {accuracy_score(y_true,y_pred)*100:.2f}%\n")

class_names = ["Glioma", "Meningioma", "Pituitary", "No Tumor"]
all_pass = True

for i, name in enumerate(class_names):
    mask = (y_true == i).astype(int)
    pred = (y_pred == i).astype(int)
    tn, fp, fn, tp = confusion_matrix(mask, pred).ravel()

    sens = tp / (tp + fn + 1e-8)
    spec = tn / (tn + fp + 1e-8)

    sens_flag = "✓" if sens >= 0.95 else "✗ BELOW TARGET"
    spec_flag = "✓" if spec >= 0.90 else "✗ BELOW TARGET"
    if sens < 0.95 or spec < 0.90:
        all_pass = False

    print(f"  {name}")
    print(f"    Sensitivity : {sens*100:.2f}%  {sens_flag}")
    print(f"    Specificity : {spec*100:.2f}%  {spec_flag}")
    print(f"    TP={tp}  FP={fp}  TN={tn}  FN={fn}\n")

print("=" * 55)
mean_sens = np.mean([(y_pred[y_true==i]==i).mean() for i in range(4)])
print(f"  Macro Sensitivity : {mean_sens*100:.2f}%")
print(f"  All targets met   : {'✓ YES' if all_pass else '✗ NOT YET'}")
print("=" * 55)

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred,
      target_names=class_names, digits=4))
```
