# Reading Images

```python
import glob
import cv2
import numpy as np

# Replace your two lists with one for each class
glioma = []
meningioma = []
pituitary = []
notumor = []

# Load glioma images
for f in glob.iglob(r"C:\Users\21ayu\Desktop\college project\MRI-brain-tumour-detection-with-gesture-control-\MRI\data\Training\glioma/*.jpg"):
    img = cv2.imread(f)
    img = cv2.resize(img,(128,128))
    b, g, r = cv2.split(img)
    img = cv2.merge([r,g,b])
    glioma.append(img)

# Load meningioma images
for f in glob.iglob(r"C:\Users\21ayu\Desktop\college project\MRI-brain-tumour-detection-with-gesture-control-\MRI\data\Training\meningioma/*.jpg"):
    img = cv2.imread(f)
    img = cv2.resize(img,(128,128))
    b, g, r = cv2.split(img)
    img = cv2.merge([r,g,b])
    meningioma.append(img)

# Load pituitary images
for f in glob.iglob(r"C:\Users\21ayu\Desktop\college project\MRI-brain-tumour-detection-with-gesture-control-\MRI\data\Training\pituitary/*.jpg"):
    img = cv2.imread(f)
    img = cv2.resize(img,(128,128))
    b, g, r = cv2.split(img)
    img = cv2.merge([r,g,b])
    pituitary.append(img)

# Load notumor images
for f in glob.iglob(r"C:\Users\21ayu\Desktop\college project\MRI-brain-tumour-detection-with-gesture-control-\MRI\data\Training\notumor/*.jpg"):
    img = cv2.imread(f)
    img = cv2.resize(img,(128,128))
    b, g, r = cv2.split(img)
    img = cv2.merge([r,g,b])
    notumor.append(img)

# Convert to numpy arrays
glioma = np.array(glioma)
meningioma = np.array(meningioma)
pituitary = np.array(pituitary)
notumor = np.array(notumor)

# Combine all images (optional)
All = np.concatenate((glioma, meningioma, pituitary, notumor))

print("All images shape:", All.shape)
```

# Visualizing brain MRI images

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_classes(glioma, meningioma, pituitary, notumor, num=5):
    classes = {
        "Glioma": glioma,
        "Meningioma": meningioma,
        "Pituitary": pituitary,
        "No Tumor": notumor
    }
    
    plt.figure(figsize=(15, 10))
    
    for row, (class_name, imgs) in enumerate(classes.items()):
        indices = np.random.choice(len(imgs), num, replace=False)
        
        for col, idx in enumerate(indices):
            plt.subplot(len(classes), num, row * num + col + 1)
            plt.imshow(imgs[idx])
            plt.title(class_name)
            plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Call it
visualize_classes(glioma, meningioma, pituitary, notumor, num=5)
```
